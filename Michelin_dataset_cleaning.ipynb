```python
import pandas as pd
```

```python
df=pd.read_csv('/Volumes/Seagate Portabl/Datasets/Kaggle/Michelin Restaurants/michelin_my_maps.csv')
df
```

```python
# check for null values in df
df.info()
df.isnull().sum()
```

```python
# convert phone numbers to strings and remove decimal place
df['PhoneNumber'] = df['PhoneNumber'].astype('str')
df['PhoneNumber'] = df['PhoneNumber'].str.strip('.0')
```

```python
df['Award'].value_counts()
```

```python
# Replace "1 Star,Green Star" with "1 Star" because that is more relavent information
df['Award'] = df['Award'].str.replace(',Green Star', '')
df['Award'].value_counts()
```

```python
df['Price'].value_counts()
```

```python
# conditions to populate new 'currency' column
pounds=(df['Price'].str.startswith('£')==True)
euros=(df['Price'].str.startswith('€')==True)
dollars=(df['Price'].str.startswith('$')==True)
yen=(df['Price'].str.startswith('¥')==True)
thai_baht=(df['Price'].str.startswith('฿')==True)
won=(df['Price'].str.startswith('₩')==True)
turkish_lira=(df['Price'].str.startswith('₺')==True)

# populate df['currency']
df.loc[pounds, 'Currency'] = 'pounds'
df.loc[euros, 'Currency'] = 'euros'
df.loc[dollars, 'Currency'] = 'dollars'
df.loc[yen, 'Currency'] = 'yen'
df.loc[thai_baht, 'Currency'] = 'thai baht'
df.loc[won, 'Currency'] = 'won'
df.loc[turkish_lira, 'Currency'] = 'turkish lira'

df['Currency'].value_counts()
```

```python
df
```

```python
# Split location into 'City' and 'Country' columns
df[['City', 'Country']] = df['Location'].str.split(',',1, expand=True)
df
```

```python
df.to_csv('michelin_updated.csv', index=False)
```
